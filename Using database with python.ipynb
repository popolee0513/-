```python
from sqlalchemy import create_engine
import pandas as pd
import pymysql
import mysql.connector
```

# 列出上海城市的商家總數

```python
engine = create_engine(str(r"mysql+mysqldb://%s:" + '%s' + "@%s/%s?charset=utf8") % ("db2017", "db2017","sqldb.demo.tornado.com.tw", "shop"))  
#(user, password, host, db)
sql ="""\
SELECT COUNT(ShopName) AS totalshop
FROM shop,city,shopcity
WHERE shop.ShopID=shopcity.ShopID
	AND shopcity.CityID=city.CityID
    AND city.CityName="上海" ; """
df = pd.read_sql_query(sql, engine)
df=pd.DataFrame(df)
df.head()
```

# 列出上海城市前十個人均單價最貴的商家

```python
sql ="""\
SELECT AvgPrice,ShopName
FROM shop,city,shopcity 
WHERE (shop.ShopID=shopcity.ShopID
	AND shopcity.CityID=city.CityID
    AND city.CityName="上海")
    ORDER BY AvgPrice+0 DESC
    LIMIT 10; """
df = pd.read_sql_query(sql, engine)
df=pd.DataFrame(df)
df
```

# 列出大陸地區前十大商家數最多的熱門菜系


```python
sql="""\
SELECT  DishName,COUNT(ShopID)
FROM   shopdish,dish
WHERE shopdish.DishID=dish.DishID
GROUP BY  DishName
 ORDER BY COUNT(ShopID)+0 DESC
    LIMIT 10;"""
df = pd.read_sql_query(sql, engine)
df=pd.DataFrame(df)
df
```

# 列出上海前十大熱門討論菜系

```python
sql="""\
SELECT DishName  ,COUNT(ShopName)
FROM shop,city,shopcity,shopdish,dish
WHERE shop.ShopID=shopcity.ShopID
	AND shopcity.CityID=city.CityID
	AND shopdish.ShopID=shop.ShopID
    AND shopdish.DishID=dish.DishID
    AND city.CityName="上海"
    GROUP BY  DishName
	ORDER BY COUNT(ShopName)+0 DESC
    LIMIT 10;
"""
df=pd.read_sql_query(sql, engine)
df=pd.DataFrame(df)
df
```

# 列出上海的川菜餐廳中，評分總分最好前20名

```python
sql="""\
SELECT  ShopName,(CommentMagnitude+CommentTaste
+CommentEnvironment+CommentService) as 'Total'
FROM shop,city,shopcity,shopdish,dish
WHERE shop.ShopID=shopcity.ShopID
	AND shopcity.CityID=city.CityID
	AND shopdish.ShopID=shop.ShopID
    AND shopdish.DishID=dish.DishID
    AND city.CityName="上海"
    AND  DishName="川菜"ORDER BY (CommentMagnitude+CommentTaste
   +CommentEnvironment+CommentService) DESC
    LIMIT 20;
"""
df=pd.read_sql_query(sql, engine)
df=pd.DataFrame(df)
df
```

# 列出大陸地區前十大總分最高的評論、作者、總分

# 列出大陸前十大總展店數最多的店家名稱與店數


```python
sql="""\
SELECT ShopName,COUNT(CityName)
FROM shop,city,shopcity 
WHERE shop.ShopID=shopcity.ShopID
	AND shopcity.CityID=city.CityID
    GROUP BY ShopName
    ORDER BY COUNT(CityName) DESC
    LIMIT 10;
"""
df=pd.read_sql_query(sql, engine)
df=pd.DataFrame(df)
df
```
