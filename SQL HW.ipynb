```python
Q1=dict()

sql="""\
SELECT count(*) 
FROM content 
WHERE title LIKE "%%好雷%%" 
AND title Like """
for i in range(len(movielist)):
    que=sql+"'"+movielist[i]+"'"+";"
    df=pd.read_sql_query(que, engine)
    df=pd.DataFrame(df)
    Q1[movielist[i]]=df.iloc[0,0]   
```

```python
Q1
```

```python
A1=pd.DataFrame(list(Q1.values()),index=title)
A1=A1.rename(columns={0: '好雷文數量'})
A1
```
